<h2>What kind of IP address is it?</h2>

```python
import ipaddress
ipaddress.ip_address('192.168.1.1')
```

```python
ipaddress.ip_address('2001:db8::1')
```

<h1>Is my IP in the network?</h1>

```python
#creating the IP network
NETWORK = ipaddress.ip_network('10.0.0.0/16')

#creating IP address object
IPV4 = ipaddress.ip_address('10.1.13.38')

print(IPV4 in NETWORK)
```

<h1>How many hosts are in the network?</h1>

```python
for host in NETWORK.hosts():
    print(host)
```
